```python
""""To design a bar chart for representing the distribution of age in the dataset.
    Bar Chart is a type of visualization for categorical data with rectangular bars, where the length or height of each bar corresponds to the frequency or proportion of the category it representsBar charts are often used to compare the values of different categories or groups visually.
    They are suitable for displaying discrete data, such as the distribution of genders, types of products sold, or the performance of different teams. """
plt.figure(figsize=(10, 12))
plt.bar(df['age'].value_counts().index, df['age'].value_counts().values, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.xticks(df['age'].unique())
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
""" To design a histogram for representing the distibution of gender("sex") in the dataset
    A histogram is used to represent the distribution of continuous data by dividing the data into intervals or bins and displaying the frequency or proportion of observations falling into each bin with adjacent rectangles.
    Histograms provide a visual representation of the underlying distribution of the data, showing the central tendency, spread, and shape of the distribution.
    They are commonly used in statistical analysis and data exploration to understand the distribution of variables, such as age distribution, income distribution, or exam scores."""
plt.figure(figsize=(9, 7))
plt.hist(df['sex'], bins=5, color='green', edgecolor='black')
plt.title('Distribution of Gender')
plt.xlabel('sex')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.9)
plt.tight_layout()
plt.show()
```
